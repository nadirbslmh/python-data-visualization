```python
import seaborn as sns
```

```python
# get datasets
sns.get_dataset_names()
```

```python
taxis = sns.load_dataset('taxis')
taxis
```

```python
sns.scatterplot(x="distance", y="total", data=taxis, hue="payment")
```

```python
sns.histplot(taxis['fare'], bins=15, kde=True)
```

```python
sns.displot(taxis['fare'], kde=True, bins=15)
```

```python
sns.barplot(x="payment", y="total", data=taxis)
```

```python
sns.boxplot(x="pickup_borough", y="total", data=taxis, hue="payment")
```

```python
sns.stripplot(x="pickup_borough", y="total", data=taxis, hue="payment", dodge=True)
```

```python
sns.jointplot(x="tip", y="total", data=taxis, kind="reg")
```

```python
sns.jointplot(x="tip", y="total", data=taxis, kind="kde", fill=True)
```

```python
sns.jointplot(x="tip", y="total", data=taxis, kind="hex")
```

```python
sns.pairplot(taxis.select_dtypes(["number"]))
```

```python
taxis_corr = taxis.corr(numeric_only=True)
taxis_corr
```

```python
sns.heatmap(taxis_corr, annot=True, cmap="coolwarm")
```

```python
cars = sns.load_dataset("mpg")
cars
```

```python
sns.catplot(
    data=cars, 
    kind="bar", 
    x="cylinders",
    y="mpg", 
    hue="origin",
)
```

```python
sns.lineplot(x="model_year", y="mpg", hue="origin", data=cars)
```

```python
health_exps = sns.load_dataset("healthexp")
health_exps
```

```python
g = sns.relplot(x="Life_Expectancy", y="Spending_USD", hue="Country", alpha=0.5, palette="muted", height=6, data=health_exps)

g.set_xlabels(label="Life Expectancy")
g.set_ylabels(label="Spending in USD")
```
