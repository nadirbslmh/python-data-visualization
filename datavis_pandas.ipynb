```python
import pandas as pd
import os
```

```python
# set the location of the dataset
%env DATASET=/home/nadir/codes/python-data-visualization/coffee_shop.csv
```

```python
dataset = os.environ["DATASET"]

df = pd.read_csv(f"{dataset}")
df.set_index("transaction_id")
df
```

```python
tx_by_date = df.groupby(["transaction_date"]).mean(numeric_only=True)["transaction_amount"]
tx_by_date.plot(kind="line")
```

```python
menu_freqs = df.groupby(["menu"]).count()["transaction_id"]
menu_freqs.plot(kind="bar", xlabel="Menu", ylabel="Frequencies", title="Ordered Menus")
```

```python
tx_by_date.hist(bins=20)
```

```python
amount_sums = df.groupby(["transaction_date"]).sum(numeric_only=True)[["transaction_amount"]]
amount_sums.boxplot()
```

```python
menu_freqs.plot.pie(y="menu", autopct="%.2f", figsize=(10, 6))
```
